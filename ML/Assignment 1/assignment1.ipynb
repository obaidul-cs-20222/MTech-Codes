```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv('/kaggle/input/housing-csv/housing.csv', header=None, delimiter=r"\s+", names=column_names)
df.head()
```

```python
df.hist (bins=50, figsize= (20,15))
```

```python
correlation_matrix=df.corr()
correlation_matrix
```

```python
    plt.figure(figsize=(8, 6)) 
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
features = df.drop('MEDV', axis=1).columns
for feature in features:
    plt.figure(figsize=(8, 5)) 
    sns.scatterplot(x=feature, y='MEDV', data=df)
    plt.title(f'MEDV vs. {feature}')
    plt.xlabel(feature)
    plt.ylabel('MEDV')
    plt.grid(True) 
    plt.show() 
```

```python
for feature in features:
    # IQR calculation for MEDV
    Q1 = df['MEDV'].quantile(0.25)
    Q3 = df['MEDV'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    
    outliers = df[(df['MEDV'] < lower_bound) | (df['MEDV'] > upper_bound)]

    
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=feature, y='MEDV', data=df, label='Normal', alpha=0.6)

    
    sns.scatterplot(x=feature, y='MEDV', data=outliers, color='red', label='Outliers', s=60, marker='X')

    plt.title(f'MEDV vs. {feature}')
    plt.xlabel(feature)
    plt.ylabel('MEDV')
    plt.grid(True)
    plt.legend()
    plt.show()
```
