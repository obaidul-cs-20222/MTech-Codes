```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
input="10101100101"
```

```python
def nrz_l(input):
    
    
    time = []
    voltage = []

    t = 0
    for bit in input:
        
        if bit == '1':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [1, 1, 1, 1]
        elif bit == '0':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [0, 0, 0, 0]
        
        t += 1
    print(f"time {time}")
    print(f"voltage {voltage}")
    
    plt.figure(figsize=(10,5))
    plt.step(time, voltage, where='post', linewidth=2)
    plt.ylim(-0.5, 1.5)
    #plt.yticks([0, 1])
    plt.xticks(range(len(input)+1))
    plt.grid(True)
    plt.title(f"NRZ-L Encoding for Binary: {input}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
def rz(input):
        
    time = []
    voltage = []

    t = 0
    for bit in input:
        
        if bit == '1':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [1, 1, 0, 0]
        elif bit == '0':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [0, 0, 0, 0]
        
        t += 1
    print(f"time {time}")
    print(f"voltage {voltage}")
    
    plt.figure(figsize=(10,5))
    plt.step(time, voltage, where='post', linewidth=2)
    plt.ylim(-0.5, 1.5)
    #plt.yticks([0, 1])
    plt.xticks(range(len(input)+1))
    plt.grid(True)
    plt.title(f"Unipolar RZ Encoding for Binary: {input}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()   
```

```python
def invert(flag):
    if flag % 2 != 0:
        voltage = [0, 0, 0, 0]
    else:
        voltage = [1, 1, 1, 1]
    return voltage

def nrz_i(input):
    

    time = [0,0]
    voltage = [1,0]
    t = 0
    flag = 0

    for bit in input:
        if bit == "1":
            flag += 1
            voltage += invert(flag)
            time += [t, t + 0.5, t + 0.5, t + 1]
        elif bit == "0":
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += invert(flag)
        t += 1

    print(f"time {time}")
    print(f"voltage {voltage}")

    
    plt.figure(figsize=(10, 5))
    plt.step(time, voltage, where='post', linewidth=2)
    plt.ylim(-1.5, 1.5)
    plt.xticks(range(len(input) + 2))
    plt.grid(True)
    plt.title(f"NRZ-I Encoding for Binary: {input}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
def polar_rz(input):
        
    time = []
    voltage = []

    t = 0
    for bit in input:
        
        if bit == '1':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [1, 1, 0, 0]
        elif bit == '0':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [-1, -1, 0, 0]
        
        t += 1
    print(f"time {time}")
    print(f"voltage {voltage}")
    
    plt.figure(figsize=(10,5))
    plt.step(time, voltage, where='post', linewidth=2)
    plt.ylim(-1.5, 1.5)
    #plt.yticks([0, 1])
    plt.xticks(range(len(input)+1))
    plt.grid(True)
    plt.title(f"Polar RZ Encoding for Binary: {input}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()  
```

```python
def manchester(input):
    
    
    time = []
    voltage = []

    t = 0
    for bit in input:
        
        if bit == '1':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [1, 1, 0, 0]
        elif bit == '0':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [0, 0, 1, 1]
        
        t += 1
    print(f"time {time}")
    print(f"voltage {voltage}")
    
    plt.figure(figsize=(10,5))
    plt.step(time, voltage, where='post', linewidth=2)
    plt.ylim(-0.5, 1.5)
    #plt.yticks([0, 1])
    plt.xticks(range(len(input)+1))
    plt.grid(True)
    plt.title(f"Manchester Encoding for Binary: {input}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```

```python
def differential_manchester(bits):
    time = []
    voltage = []

    t = 0
    
    current_level = 1  

    for bit in bits:
        if bit == '0':
            current_level = 1 - current_level  
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [current_level, current_level, 1 - current_level, 1 - current_level]

        elif bit == '1':
            
            time += [t, t + 0.5, t + 0.5, t + 1]
            voltage += [current_level, current_level, 1 - current_level, 1 - current_level]

        
        current_level = 1 - current_level  
        t += 1

    print(f"time {time}")
    print(f"voltage {voltage}")

    plt.figure(figsize=(10, 5))
    plt.step(time, voltage, where='post', linewidth=2)
    plt.ylim(-0.5, 1.5)
    plt.xticks(range(len(bits)+1))
    plt.grid(True)
    plt.title(f"Differential Manchester Encoding for Binary: {bits}")
    plt.xlabel("Time")
    plt.ylabel("Voltage Level")
    plt.show()
```
